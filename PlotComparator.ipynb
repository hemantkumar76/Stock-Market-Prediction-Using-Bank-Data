```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
actual = pd.read_csv("YESBANK.NS.csv")['Close']
vae_elm = np.load('VAE_ELM.npy')
ae_elm = np.load('AE_ELM.npy')
rbf = np.load('RBF.npy')
```

```python
fig, axs = plt.subplots(1, 1, figsize=(9,6))
axs.plot(actual, label='Original Data', color='r')
axs.plot(vae_elm, label='VAE ELM', linestyle='-.')
axs.plot(ae_elm, label='AE ELM', linestyle='--')
axs.plot(rbf, label='RBF', linestyle=':')
axs.legend()
axs.grid()
plt.savefig('Predicted Closing Price')
plt.show()
```
